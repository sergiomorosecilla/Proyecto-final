```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

1.Procedo a cargar los datos del archivo CSV y realizo algunas modificaciones en las columnas. Despues de ell mostramos resumen de la info.

```python
#Cargo el archivo csv
df_bank = df_bank_additional = pd.read_csv('./bank-additional.csv')
```

```python
#Me aseguro que las columnas tengan nombres en min√∫sculas
df_bank.columns = df_bank_additional.columns.str.lower()
# Elimino la columna 'unnamed: 0' que es innecesaria y otras 3 que no
#utilizaremos en nuestro an√°lisis, ya que nos centraremos en resultados de campa√±a de marketing
drops = ['unnamed: 0', 'cons.price.idx', 'cons.conf.idx', 'euribor3m']
df_bank.drop(columns = drops , axis=1, inplace=True)
#Renombro la columna 'y' a 'subscribed' para mayor claridad
df_bank.rename(columns={'y': 'subscribe'}, inplace=True) 
#Renombro la columna 'id_' por 'id' para poder hacer merge con el otro dataset a trav√©s de esta columna
df_bank.rename(columns={'id_': 'id'}, inplace=True)
# Verifico que no haya duplicados
print(df_bank.duplicated('id').sum())                                                                                 
#Obtenemos la info sobre la cantidad de datos (columnas y filas) y tipos de variables 
df_bank.info() 
#Mostramos un resumen de las primeras filas y columnas del archivo
df_bank.head()
```

2. Realizo lo mismo para el archivo XLSX.

```python
#Procedo con los mismos pasos para el archivo xlsx
# Leer todas las hojas del Excel
excel_path = 'customer-details.xlsx'
all_sheets = pd.read_excel(excel_path, sheet_name=None)

# Unir todas las hojas en un solo DataFrame
df_customer = df_customer_details = pd.concat(all_sheets.values(), ignore_index=True)

# Verificar duplicados en IDs
print(df_customer['ID'].duplicated().sum())
```

```python
#Me aseguro que las columnas tengan nombres en min√∫sculas
df_customer.columns = df_customer.columns.str.lower()
# Elimino columna 'unnamed: 0'que es innecesaria
df_customer.drop(columns ='unnamed: 0' , axis=1, inplace=True)
#Renombro la columna 'ID' por 'id' para poder hacer merge con el otro dataset
df_customer.rename(columns={'ID': 'id'}, inplace=True)                                                                                       
#Obtenemos la info sobre la cantidad de datos (columnas y filas) y tipos de variables 
df_customer.info() 
#Mostramos un resumen de las primeras filas y columnas del archivo
df_customer.head()
```

3. Uno los dos datasets a trav√©s de sus respectivas columnas 'id' que es la referencia de las clientes y puede servir como clave √∫nica. Acabar√© de realizar limpieza final de datos

```python
# Realizar outer join (todos los registros de ambos DataFrames)
df_merged = df_bank.merge(df_customer, on='id', how='outer')
# Tama√±os
print("Tama√±o df_bank:", df_bank.shape)
print("Tama√±o df_customer:", df_customer.shape)
print("Tama√±o df_merged (outer):", df_merged.shape)

df_merged.info()
df_merged.head()
```

Realizo √∫ltimas operaciones de limpieza o mejora de visualizaci√≥n de datos

```python
#Mejoro visibilidad de variables buleanas 'default','housing' y 'loan' para que muestren
#un valor booleano limpio sin decimales y en datos vac√≠os o NaN muestre un 0.
cols = ['default', 'housing', 'loan']
# Reemplazar valores vac√≠os o NaN con 0 y convertir todos los valores a enteros
df_merged[cols] = df_merged[cols].replace({'': 0}).fillna(0).astype(int)
# En columna date:
# Renombro columna por call_date para mayor claridad
df_merged.rename(columns={'date': 'call_date'}, inplace=True)
# Convertir la columna 'call_date' a tipo datetime para facilitar an√°lisis posteriores
# para ello he utilizado la librer√≠a locale y una funci√≥n de pandas, que he buscado en ChatGPT
import locale
locale.setlocale(locale.LC_TIME, 'Spanish_Spain')  # Windows (alternativa)
df_merged['call_date'] = pd.to_datetime(df_merged['call_date'], format='%d-%B-%Y', errors='coerce')


# Verificar resultado
df_merged.info()
#Hecho un vistazo a la cantidad de valores nulos
df_merged.isnull().sum()
```
