```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

1.Procedo a cargar los datos del archivo CSV y realizo algunas modificaciones en las columnas

```python
df_btc = pd.read_csv('BTC.csv')

df_btc.info()
df_btc.head()

```

```python
# Calcular variación porcentual diaria y añadirla como nueva columna
df_btc["%var_btc"] = (((df_btc["close"] - df_btc["open"]) / df_btc["open"]) * 100).round(2)
# Cambiar el formato de la columna 'date' que será con la que haremos merge con los otros datos
df_btc['date'] = pd.to_datetime(df_btc['date'], errors='coerce', dayfirst=True)
# Cambiar el formato de la columna 'volume' a int64 ya que es un valor que puede tomar altos valores y no es decimal
df_btc["volume"] = df_btc["volume"].astype('int64')
# Eliminamos la columna 'tiker' que en nuestro set ya viene identificado en columna
df_btc.drop(columns = 'ticker', inplace=True)

df_btc.info()
df_btc.head()
```

```python
#Seguimos ahora con carga datos y cambios sobre datos S&P500
df_sp = pd.read_csv('sp-historical.csv')

df_sp.info()
df_sp.head()
```

```python
# Pasar a minúsculas los nombres de las columnas
df_sp.columns = df_sp.columns.str.lower()
# Nueva columna con la variación porcentual diaria
df_sp["%var_sp"] = (((df_sp["close"] - df_sp["open"]) / df_sp["open"]) * 100).round(2)
# Eliminar columna 'adj close' (redundante)
df_sp.drop(columns = 'adj close', inplace=True)
# Cambiar formato de la columna 'date' a datetime
df_sp['date'] = pd.to_datetime(df_sp['date'], errors='coerce', dayfirst=True)


df_sp.info()
df_sp.head()
```

```python
# Último set a cargar y modificar, datos de cotización del oro
df_oro = pd.read_csv('fut_oro.csv')

df_oro.info()
df_oro.head()
```

```python
# Normalizamos nombre columnas a semejanza de los otros 2 datasets
df_oro.rename(columns={'Fecha': 'date', 'Último': 'close_gd', 'Apertura': 'open_gd', 'Máximo': 'high_gd', 'Mínimo': 'low_gd', 'Vol.': 'volume_gd', '% var.':'%var_gd'}, inplace=True)

##Antes de converir a número entero, limpiamos símbolos y demás:
#1 Primero quitamos símbolos como comas, puntos o 'K' que puedan venir en texto
#2 Eliminamos comas de miles y el sufijo 'K' en volumen
df_oro["volume_gd"] = (df_oro["volume_gd"].astype(str).str.replace("K", "000", regex=False)   # convierte 9,30K → 9300
    .str.replace(",", "", regex=False)      # quita comas
)
#3 Convertir volumen a int64, usando 'coerce' para convertir errores a NaN y luego a Int64 que acepta NaN
df_oro["volume_gd"] = pd.to_numeric(df_oro["volume_gd"], errors="coerce").astype("Int64")

# Convertir otras columnas a float
# Los valores vienen con formato español: punto como separador de miles y coma como decimal (ej. '3.768,10')
# Primero normalizamos (quitar separador de miles '.' y cambiar ',' por '.') y luego convertimos a numerico
cols_float = ["close_gd", "open_gd", "high_gd", "low_gd"]
df_oro[cols_float] = (
    df_oro[cols_float]
    .astype(str)
    .replace({r'\.': '', ',': '.'}, regex=True)   # 3.768,10 -> 3768.10
    .apply(pd.to_numeric, errors="coerce")
)

# Quitar de la columna '%var_gd' símbolo % y posibles comas decimales y despues convertir a float:
df_oro["%var_gd"] = (
    df_oro["%var_gd"]
    .astype(str)
    .str.replace("%", "", regex=False)     
    .str.replace(",", ".", regex=False) 
)
df_oro["%var_gd"] = df_oro["%var_gd"].astype(float)

df_oro['date'] = pd.to_datetime(df_oro['date'])


df_oro.info()
df_oro.head()
```

```python
# Realizar merges de forma secuencial:
df_merged = df_btc.merge(df_sp, on='date', how='inner', suffixes=('_btc', '_sp'))
df_merged = df_merged.merge(df_oro, on='date', how='inner', suffixes=('', '_oro'))

print(f'Tamaño: df_btc.shape, {df_btc.shape}')
print(f'Tamaño: df_sp.shape, {df_sp.shape}')
print(f'Tamaño: df_oro.shape, {df_oro.shape}\n')

df_merged.info()
df_merged.isnull().sum()
```
