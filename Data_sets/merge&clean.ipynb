```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
df_btc = pd.read_csv('datos_BTC_USD.csv')

df_btc.info()
df_btc.head()
df_btc.tail()
```

```python
df_sp = pd.read_csv('datos_S&P500.csv')

df_sp.info()
df_sp.head()
df_sp.tail()
```

```python
df_oro = pd.read_csv('datos_futuros_oro.csv')

df_oro.info()
df_oro.head()
```

```python
# Realizar outer join (todos los registros de ambos DataFrames)
# merge no acepta una tupla: hacer merges de forma secuencial
df_merged = df_btc.merge(df_sp, on='Fecha', how='inner', suffixes=('_btc', '_sp'))
df_merged = df_merged.merge(df_oro, on='Fecha', how='inner', suffixes=('', '_oro'))
# Tamaños

df_merged.info()
df_merged.head()
```
